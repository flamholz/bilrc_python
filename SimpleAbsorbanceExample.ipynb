```python
import numpy as np
import seaborn as sns
import pandas as pd

from sklearn.metrics import r2_score
from matplotlib import pyplot as plt
from util.bilrc_util import read_ns1_data
```

```python
# Load absorbance data from binary files
abs_data = read_ns1_data('data/10_19_2020_Vinay_dimer_p56_TAS_11.ns1')

# read_ns1_data returns an object that can be pretty-printed.
print(abs_data)

# Resample the data to have logarithmically-spaced timepoints
logtime_abs_data = abs_data.log_time()
```

```python
# Plot the absorbance as a function of time.
figure = plt.figure()
plt.plot(logtime_abs_data.T*1e6, logtime_abs_data.Y)
plt.xlabel('Time ($\mu s$)')
plt.ylabel('Absorbance')

plt.show()
```
